```python
import numpy as np
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('Image1.jpg')
```

```python
img2 = cv2.imread('Image2.jpg')
```

```python
img3 = cv2.imread('Image3.jpeg')
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray1, cmap = 'gray')
```

```python
plt.imshow(gray2, cmap = 'gray')
```

```python
plt.imshow(gray3, cmap = 'gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces_reads1 = haar_cascade_face.detectMultiScale(gray1, scaleFactor = 1.2, minNeighbors = 5)
```

```python
faces_reads2 = haar_cascade_face.detectMultiScale(gray2, scaleFactor = 1.2, minNeighbors = 5)
```

```python
faces_reads3 = haar_cascade_face.detectMultiScale(gray3, scaleFactor = 1.2, minNeighbors = 5)
```

```python
print('Faces Found:', len(faces_reads1))
```

```python
print('Faces Found:', len(faces_reads2))
```

```python
print('Faces Found:', len(faces_reads3))
```

```python
for(m, n, x, y) in faces_reads1:
    cv2.rectangle(img1, (m,n), (m+x, n+y), (0, 255, 0), 2)
```

```python
for(m, n, x, y) in faces_reads2:
    cv2.rectangle(img2, (m,n), (m+x, n+y), (0, 255, 0), 2)
```

```python
for(m, n, x, y) in faces_reads3:
    cv2.rectangle(img3, (m,n), (m+x, n+y), (0, 255, 0), 2)
```

```python
def convert2RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(convert2RGB(img1))
```

```python
plt.imshow(convert2RGB(img2))
```

```python
plt.imshow(convert2RGB(img3))
```
